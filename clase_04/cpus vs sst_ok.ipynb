## SST vs CPUE

```python
import numpy as np
import os 
os.getcwd()
```

```python
# cpue
data=np.load('cpueGuatemala_mea.npz')
```

```python
#data.files #para encontrar las claves de la estructura
cpue=data['media']
```

```python
# sst
data2=np.load('mediaSST guatemala.npz')
data2.files
sst3d=data2['mediaSST']
```

```python
mediaSST=[]
for ii in range(0,12):
    mediaSST.append(np.nanmean(sst3d[ii,:,:]))

mediaSST=np.array(mediaSST)
#print(mediaSST)
```

```python
mediaSST.shape
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
labels=['Enero','Febrero','Marzo','Abril']

dataSST=mediaSST[0:4]

fig,ax1=plt.subplots()
ax1.plot(x,cpue)
ax1.tick_params(axis='y',color='C0',labelcolor='C0')
ax1.set_ylabel('CPUE',color='C0')

ax2=ax1.twinx()
ax2.plot(x,dataSST,'C1')
ax2.tick_params(axis='y',color='C1',labelcolor='C1')
ax2.set_ylabel('Temperature',color='C1')

plt.xticks(x,labels)
plt.show()
```
