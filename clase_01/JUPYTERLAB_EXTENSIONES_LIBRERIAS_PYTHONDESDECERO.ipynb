
# Python
### Python es un lenguaje de programación de alto nivel que se utiliza para desarrollar aplicaciones de todo tipo. A diferencia de otros lenguajes como Java o .NET, se trata de un lenguaje interpretado, es decir, que no es necesario compilarlo para ejecutar las aplicaciones escritas en Python, sino que se ejecutan directamente por el ordenador utilizando un programa denominado interpretador, por lo que no es necesario “traducirlo” a lenguaje máquina.

# JupyterLab
### JupyterLab es el último entorno de desarrollo interactivo basado en la web para cuadernos, código y datos. Su interfaz flexible permite a los usuarios configurar y organizar flujos de trabajo en ciencia de datos, computación científica, periodismo computacional y aprendizaje automático. Un diseño modular invita a las extensiones para ampliar y enriquecer la funcionalidad.



https://jupyter.org/ 
***

## EXTENSIONES
### Fundamentalmente, JupyterLab está diseñado como un entorno extensible. Las extensiones de JupyterLab pueden personalizar o mejorar cualquier parte de JupyterLab. Pueden proporcionar nuevos temas, visores y editores de archivos, o renderizadores para resultados enriquecidos en cuadernos.  De hecho, todo JupyterLab en sí mismo es simplemente una colección de extensiones que no son más poderosas ni privilegiadas que cualquier extensión personalizada.
https://jupyterlab.readthedocs.io/en/latest/user/extensions.html

## LIBRERIAS
### Las librerías de Python son conjuntos de funciones que permiten realizar distintas funciones, ahorrando tiempo y esfuerzo al programador. Una vez que las librerías o bibliotecas están instaladas en tu computadora, puedes utilizarlas importaándolas al principio de tu código. Puedes importar tantas librerías como quieras

* import NOMBRE_DE_LA_LIBRERIA
* from NOMBRE_DE_LA_LIBRERIA import MODULO
