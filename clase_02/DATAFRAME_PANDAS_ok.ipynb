```python
%reset -f
import pandas as pd
import numpy as np
```

```python
file='CPUE PACIFICO GUATEMALA.xlsx'
df=pd.read_excel(file)
```

```python
print(df)
```

```python
mes=np.array(df['MES'])
year=np.array(df['AÑO'])
cpue=np.array(df['CPUE'])
```

```python
mes2=[1,2,3,4]
media_cpue=np.zeros(4)
for i in range(0,4):#
    ind=np.where(mes==mes2[i])# para calcular los indices de los meses
    media_cpue[i]=np.mean(cpue[ind])    
print(media_cpue)  
```

```python
# para hacer la figura
dicc2={'meses':['Enero','Febrero','Marzo','Abril'],'Media_Cpue': media_cpue}
df2=pd.DataFrame(dicc2)
```

```python
df2.plot(x='meses', y='Media_Cpue',kind='bar')
```
